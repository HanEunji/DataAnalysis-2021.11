- 결측치 처리

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.info()
```

```python
titanic.isna().sum().sum() # 전체 결측치 확인
```

```python
titanic.isna().sum()
```

```python
titanic.tail()
```

```python
# age 결측치를 평균으로 대체
titanic['age'].fillna(titanic.age.mean(), inplace=True)
titanic.tail(3)
```

```python
# embarked 결측치는 행 삭제
titanic.dropna(subset=['embarked'], inplace=True)
titanic.dropna(subset=['embark_town'], inplace=True)
titanic.tail()
```

```python
# deck 컬럼은 컬럼 
titanic.dropna(columns=['deck'], inplace=True)
titanic
```

- apply

```python
# 각 열의 최대값과 최소값의 차이
df.apply(lambda x: x.max() - x.min()) # axis = 0 생략
```

```python
# 각 행의 최대값과 최소값의 차이
df.apply(lambda x: x.max() - x.min(), axis=1)
```

```python
# 타이타닉호 승객 중 20살이 넘으면 'adult' 그렇지 않으면 'child'
titanic['adult/child'] = titanic.age.apply(lambda x: 'adult' if x > 20 else 'child')
    # titanic.apply(lambda x: 'adult' if x.age > 20 else 'child', axis=1)
titanic.tail()
```

```python
titanic['ac'] = titanic.age.apply(lambda x: 'adult' if x > 20 else 'child')
titanic.tail()
```

```python
# cctv 소수점없애기(?) 숫자 구분자(,) 없애기.
cctv.head()
```

```python
cctv['총계'] = cctv['총계'].apply(lambda x: int(x.replace(',',''))) # '77,032'라는 문자열을 그냥 인트로 변환하면 오류남. 콤마(,) 지우고 변환
cctv
```

```python
cctv.columns
```

```python
for column in cctv.columns[2:]:
    cctv[column] = cctv[column].apply(lambda x: int(x.replace(',','')))
cctv.head()
```
