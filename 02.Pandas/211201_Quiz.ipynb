```python
# 1-b
# Q1 - 1.5 * (Q3 - Q1)보다 작은 데이터, Q3 + 1.5 * (Q3 - Q1)보다 큰 데이터는 이상치
s = iris[iris.species == 'setosa']['sepal_length']
Q1 = s.quantile(0.25)
Q3 = s.quantile(0.75)
```

```python
# 이상치 제거
lo = Q1 - 1.5 * (Q3 - Q1)
up = Q3 + 1.5 * (Q3 - Q1)
s[(s<lo)|(s>up)] = np.nan
# 종별로 꽃잎길이, 꽃잎폭, 꽃받침길이, 꽃받침폭의 평균과 표준편차 구하기
s.mean(), s.std()
```

```python
def iqr(sp, co):
    s = iris[iris.species == sp][co]
    Q1 = s.quantile(0.25)
    Q3 = s.quantile(0.75)
    lo = Q1 - 1.5 * (Q3 - Q1)
    up = Q3 + 1.5 * (Q3 - Q1)
    s[(s<lo)|(s>up)] = np.nan
    return s.mean(), s.std()

iqr('versicolor', 'sepal_width')
```
