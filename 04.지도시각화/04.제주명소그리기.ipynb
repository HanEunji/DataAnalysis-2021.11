# 제주명소 그리기

```python
import folium
import pandas as pd
from folium.features import CustomIcon
```

```python
df = pd.read_csv('제주명소2.csv')
df
```

```python
map = folium.Map(location=[df.위도.mean(), df.경도.mean()], zoom_start=10.5)
for i in df.index:
    folium.Marker(location=[df.위도[i], df.경도[i]], tooltip=df.명소[i], popup=folium.Popup(df.도로명주소[i], max_width=200)).add_to(map)
title = '<h3 align="center" style="font-size:20px">제주 주요명소 10개</h3>'
map.get_root().html.add_child(folium.Element(title))
map
```

```python
map.save('제주명소.html')
```

```python
map = folium.Map(location=[df.위도.mean(), df.경도.mean()], zoom_start=10.5)

icon_image1 = 'C:/Workspace/02.DataAnalysis/04.지도시각화/holitree.png'
icon1 = folium.features.CustomIcon(icon_image1,
    icon_size=(75, 95))

for i in df.index:
    marker = folium.Marker(
        location=[df.위도[i], df.경도[i]],
        icon=icon1,
        tooltip=df.명소[i],
        popup=folium.Popup(df.도로명주소[i], max_width=200))

map.add_child(marker)

title = '<h3 align="center" style="font-size:20px">제주 주요명소 10개</h3>'
map.get_root().html.add_child(folium.Element(title))

map
```
