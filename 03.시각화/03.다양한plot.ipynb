```python
X, Y = np.random.randn(100), np.random.randn(100)
plt.title('산점도')
plt.scatter(X, Y)
plt.grid()
plt.show()
```

### 2. Bar Chart

##s 4. Box plot

```python
# setosa
iris = sns.load_dataset('iris')
setosa = iris[iris.species == 'setosa']
```

```python
plt.figure(figsize=(8,10))
plt.boxplot([setosa.sepal_length, setosa.sepal_width,
            setosa.petal_length, setosa.petal_width],
            labels=iris.columns[:-1])
plt.title('Setosa 품종의 각 피쳐', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(4,6))
plt.boxplot(setosa.petal_width)
plt.title('Setosa petal_width', fontsize=15)
plt.show()
```

### 5. 파이(Pie) 차트

```python
ratio = [36, 32, 16, 18]
labels = ['사과', '바나나', '멜론', '포도']

plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.show() # 기본은 3시 방향에서 시작해 반시계방향으로 돌아감
```

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```
