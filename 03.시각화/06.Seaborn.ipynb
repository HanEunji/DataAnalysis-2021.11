# 데이터 시각화 - Seaborn

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
# plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.family'] = 'Pretendard-sb'
```

### 1. Count plot

```python
titanic = sns.load_dataset('titanic')
sns.countplot(x='class', data=titanic)
plt.title('타이타닉호의 각 클래스별 승객수')
plt.show()
```

```python
sns.countplot(x='embark_town', data=titanic);
```

# 2. Pair plot

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
```

### 3. Heatmap

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights.shape
```

```python
passengers = flights.pivot('month', 'year', 'passengers') # 행, 열, 데이터
passengers.head()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(passengers, annot=True, fmt='d', linewidths=1)
plt.show()
```

### 4. Box plot

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('요일별 전체요금')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', hue='sex', data=tips)
plt.title('요일별/성별 전체요금')
plt.show()
```

```python
iris = sns.load_dataset('iris')
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.show()
```

```python
sns.boxplot(data=iris[iris.species=='setosa'])
```
