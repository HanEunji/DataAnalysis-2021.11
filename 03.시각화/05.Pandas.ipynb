# Pandas 시각화

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
# plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.family'] = 'Pretendard-sb'
```

### 1. Line chart

```python
np.random.seed(2021)
df = pd.DataFrame(np.random.randn(100,3),
index=pd.date_range('1/1/2021', periods=100),
columns=['A','B','C']).cumsum()

df.plot() # 라인그래프를 그리겠다
plt.title('Pandas의 plot method 사용 예', loc='right', pad=15, fontsize=20)
plt.xlabel('일자'), plt.ylabel('데이터')
plt.show()
```

```python
df.head() # 인덱스(일자)가 x 좌표
```

### 2. Bar chart

```python
import seaborn as sns
iris = sns.load_dataset('iris')

plt.figure(figsize=(12,6))
iris.sepal_length[:50].plot(kind='bar', rot=45) # 앞에서부터 50개 뽑으면 세토사 품종 # rot=0로 인덱스 글자 방향 돌림
plt.title('Setosa 품종의 Sepal length')
plt.show()
```

```python
df2 = iris.groupby('species').mean()
df2.columns.name = 'Feature'
df2
```

```python
df2.plot.bar(rot=0)
```

### 3.Pie chart

```python
titanic = sns.load_dataset('titanic')
s = titanic['class'].value_counts()
s
```

```python
s.plot.pie(autopct='%.2f%%')
plt.title('선실별 승객수')
plt.show()
```

### 4. Box plot

```python
iris[iris.species == 'setosa'].plot.box(figsize=(8,10))
plt.title('Setosa 품종의 각 Feature별 Box plot', fontsize=15)
plt.show()
```

```python
iris.boxplot(by='species', figsize=(10,8))
plt.title('각 피쳐의 품종별 데이터에 대한 Box plot')
# iris.suptitle("각 피쳐의 품종별 데이터에 대한 Box plot", fontsize=50)
plt.show()
```

```python
axes = iris.boxplot(by='species', figsize=(10,8))

fig = axes[0][0].get_figure()
fig.suptitle('각 피쳐의 품종별 데이터에 대한 Box plot', color='red', ha='left', va='bottom', fontsize=20)
plt.show()
```
