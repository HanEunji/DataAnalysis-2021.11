# Matplotlib - 여러개의 그래프

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
# plt.rcParams['font.family'] = 'Malgun Gothic'
# plt.rcParams['font.family'] = 'Pretendard'
plt.rcParams['font.family'] = 'Pretendard-sb'
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
def g(t):
    return np.sin(np.pi*t)
t1 = np.arange(0.0, 5.0, 0.01)
t2 = np.arange(0.0, 5.0, 0.01)

plt.subplot(221) # 2 x 2, 그래프중 첫번째
plt.plot(t1, f(t1))
plt.subplot(2, 2, 2)
plt.plot(t2, g(t2))
plt.subplot(223)
plt.plot(t1, f(t1), 'r-')
plt.subplot(224)
plt.plot(t2, g(t2), 'r-')
plt.show()
```

- Anscombe's Quartet

```python
import seaborn as sns
ans = sns.load_dataset('anscombe')
ans.head(3)
```

```python
ds = ans[ans.dataset == 'I']
plt.scatter(ds.x, ds.y, marker='o')
plt.title('Dataset I')
plt.show()
```

```python
# 하나를 선택해서 연습
# y = 0.5x + 3
ds = ans[ans.dataset == 'I']
plt.scatter(ds.x, ds.y, marker='o')
plt.plot([4,14], [5,10], 'r-', lw=2)
plt.title('Dataset I')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10,8))
for index, roman in enumerate(ans.dataset.unique()):    # 혹은 ['I', 'II', 'III', 'IV']
    ds = ans[ans.dataset == roman]
    ax = axes[index//2, index%2]
    ax.scatter(ds.x, ds.y, marker='o')
    ax.plot([3,20], [4.5,13], 'r-', lw=2)
    ax.set_title(f'Dataset {roman}')

fig.suptitle("Anscombe's Quartet", fontsize=50)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10,8), sharex=True, sharey=True)
for index, roman in enumerate(ans.dataset.unique()):    # 혹은 ['I', 'II', 'III', 'IV']
    ds = ans[ans.dataset == roman]
    ax = axes[index//2, index%2]
    ax.scatter(ds.x, ds.y, marker='o')
    ax.plot([3,20], [4.5,13], 'r-', lw=1)
    ax.set_title(f'Dataset {roman}')

fig.suptitle("Anscombe's Quartet", fontsize=50)
plt.tight_layout()
plt.show()
```
