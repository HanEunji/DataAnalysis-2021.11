# Matplotlib - plot 기본

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
# plt.rcParams['font.family'] = 'Malgun Gothic'
# plt.rcParams['font.family'] = 'Pretendard'
plt.rcParams['font.family'] = 'Pretendard-sb'
```

```python
plt.plot([0,1,2,3],[0,1,4,9])
plt.title('라인 플롯')
plt.grid()
plt.show()
```

```python
plt.plot([0,1,2,3],[0,1,4,9], 'ro-') # color, marker, linestyle
plt.title('Red-Circle-Line')
plt.grid()
plt.show()
```

```python
plt.plot([0,1,2,3],[0,1,4,9], 'ms--') # color, marker, linestyle
plt.title('Magenta-Square-Dashed')
plt.grid()
plt.show()
```

```python
plt.plot([0,1,2,3],[0,1,4,9], 'ms--') # color, marker, linestyle
plt.plot([8,4,2,1], 'g^--')
plt.title('Magenta-Square-Dashed, Green-삼각형-점선')
plt.grid()
plt.show()
```

```python
plt.plot([0,1,2,3],[0,1,4,9], 'ms--', linewidth=3, markersize=10) # color, marker, linestyle
plt.plot([8,4,2,1], 'g^--', lw=5, ms=20)
plt.title('Magenta-Square-Dashed, Green-삼각형-점선')
plt.grid()
plt.show()
```

- 그림 범위

```python
plt.plot([0,1,2,3],[0,1,4,9], 'ro-')
plt.title('x축, y축 범위 지정')
plt.xlim(-1, 4)
plt.ylim(-1,10)
plt.grid()
plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 500)
Y = np.cos(X)
plt.title('Tick 설정')
plt.plot(X,Y)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.grid()
plt.show()
```

- 범례(Legend)

```python
X = np.linspace(-np.pi, np.pi, 500)
C, S = np.cos(X), np.sin(X)
plt.title('Legend를 표시한 플롯')
plt.plot(X, C, label='cosine')
plt.plot(X, S, label='sine')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.grid()
plt.legend()
plt.show()
```

- x축, y축 Label

```python
X = np.linspace(-np.pi, np.pi, 500)
C, S = np.cos(X), np.sin(X)
plt.title('X축, Y축 라벨을 표시한 플롯')
plt.plot(X, C, label='cosine')
plt.plot(X, S, label='sine')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.grid(), plt.legend()
plt.xlabel('time'), plt.ylabel('Amplitude')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.title('그림 크기 지정')
plt.plot(X, C, label='cosine')
plt.plot(X, S, label='sine')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.grid(), plt.legend()
plt.xlabel('time'), plt.ylabel('Amplitude')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.title('그림 크기 지정한 그래프를 저장')
plt.plot(X, C, label='cosine')
plt.plot(X, S, label='sine')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.grid(), plt.legend()
plt.xlabel('time'), plt.ylabel('Amplitude')
plt.savefig('삼각함수.png')
```
